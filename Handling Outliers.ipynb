```python
outliers = data[(data["Height(Inches)"] > 73.69) | (data["Height(Inches)"] < 62.28)]
outliers
```

```python
data_no_outlier = data[(data["Height(Inches)"] < 73.69) & (data["Height(Inches)"] > 62.28)]
data_no_outlier
```

## Outlier Removal using Z-score

```python
data['zscore'] = (data["Height(Inches)"] - data["Height(Inches)"].mean())/data["Height(Inches)"].std()
```

```python
data
```

```python
data[data['zscore'] < -3]
```

```python
data[data['zscore'] > 3]
```

```python
# Total Outliers using Z SCORE
outlier_z = data[(data['zscore']>3) |( data['zscore']<-3)]
print(outlier_z)
outlier_z.shape
```

```python
data_no_outlier_z = data[(data['zscore']>-3)&(data['zscore']<3)]
data_no_outlier_z
```
